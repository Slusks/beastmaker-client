```python
import random

def calculate_speed_modifier(monster_a, monster_b):
    return 2 * (monster_a['Speed'] - monster_b['Speed'])

def calculate_strength_modifier(monster_a, monster_b):
    return 5 * abs(monster_a['Strength'] - monster_b['Strength'])

def calculate_cunning_modifier(monster_a, monster_b):
    return 10 + 2 * (monster_a['Cunning'] - monster_b['Cunning'])

def calculate_tenacity_modifier(monster_a, monster_b):
    return 15 if monster_a['Tenacity'] > monster_b['Tenacity'] else 0

def process_roll(roll, monster_a, monster_b, strength_modifier, cunning_modifier, strength_rerolls, cunning_steals):
    """
    Processes a dice roll, checking it against Strength and Cunning modifiers.
    """
    # Strength reroll mechanic
    if strength_modifier > 0 and roll in [3, 4]:
        print(f"ShadowDrake lost on a {roll}. Checking for reroll...")
        new_roll = random.randint(1, 6)
        if new_roll > 3:
            print(f"Strength ability activated! Reroll: {new_roll}")
            strength_rerolls.append(1)  # Log reroll occurrence
            return monster_a['name']  # Reroll succeeds for Monster A
        else:
            return monster_b['name']  # Reroll fails, point remains with Monster B
    
    # Cunning steal mechanic
    if cunning_modifier > 0 and roll not in [1, 6]:
        if random.random() < (cunning_modifier / 100):
            cunning_steals.append(1)  # Log steal occurrence
            return monster_a['name']  # Cunning allows Monster A to steal the roll
    
    return monster_a['name'] if roll <= 3 else monster_b['name']

def battle_simulation(monster_a, monster_b):
    """
    Simulates a battle between two monsters by rolling a six-sided die 601 times.
    Monster A wins on rolls of 1-3, Monster B wins on rolls of 4-6.
    Includes Strength and Cunning mechanics.
    """
    scores = {monster_a['name']: 0, monster_b['name']: 0}
    total_rolls = 200
    
    strength_modifier = calculate_strength_modifier(monster_a, monster_b)
    cunning_modifier = calculate_cunning_modifier(monster_a, monster_b)
    
    strength_rerolls = []
    cunning_steals = []
    
    for _ in range(total_rolls):
        roll = random.randint(1, 6)
        winner = process_roll(roll, monster_a, monster_b, strength_modifier, cunning_modifier, strength_rerolls, cunning_steals)
        scores[winner] += 1
    
    print("\nBattle Summary:")
    print(f"Total Strength Rerolls: {len(strength_rerolls)}")
    print(f"Total Cunning Steals: {len(cunning_steals)}")
    
    return scores

# Example input for the beasts
monster_a = {"name": "ThunderBeast", "Speed": 7, "Strength": 2, "Cunning": 7, "Tenacity": 6}
monster_b = {"name": "ShadowDrake", "Speed": 5, "Strength": 4, "Cunning": 7, "Tenacity": 7}

# Calculate attribute modifiers
speed_modifier = calculate_speed_modifier(monster_a, monster_b)
strength_modifier = calculate_strength_modifier(monster_a, monster_b)
cunning_modifier = calculate_cunning_modifier(monster_a, monster_b)
tenacity_modifier = calculate_tenacity_modifier(monster_a, monster_b)

# Print attribute modifiers
print("Attribute Modifiers:")
print(f"Speed Modifier: {speed_modifier}")
print(f"Strength Modifier: {strength_modifier}")
print(f"Cunning Modifier: {cunning_modifier}")
print(f"Tenacity Modifier: {tenacity_modifier}")

# Run battle simulation
battle_result = battle_simulation(monster_a, monster_b)

# Print final scores
print("\nFinal Scores:")
print(battle_result)
```

```python
import random

def calculate_speed_modifier(monster_a, monster_b):
    return 2 * (monster_a['Speed'] - monster_b['Speed'])

def calculate_strength_modifier(monster_a, monster_b):
    return 5 * abs(monster_a['Strength'] - monster_b['Strength'])

def calculate_cunning_modifier(monster_a, monster_b):
    return 10 + 2 * (monster_a['Cunning'] - monster_b['Cunning'])

def calculate_tenacity_modifier(monster_a, monster_b):
    return 15 if monster_a['Tenacity'] > monster_b['Tenacity'] else 0

def process_roll(roll, monster_a, monster_b, speed_modifier, cunning_modifier, strength_rerolls, cunning_steals):
    if roll <= 3 and speed_modifier > 0:
        if random.random() < (speed_modifier / 100):
            return monster_a['name']
    elif roll > 3 and speed_modifier < 0:
        if random.random() < (abs(speed_modifier) / 100):
            return monster_b['name']
    
    if cunning_modifier > 0 and roll not in [1, 6]:
        if random.random() < (cunning_modifier / 100):
            cunning_steals.append(1)
            return monster_a['name']
    
    return monster_a['name'] if roll <= 3 else monster_b['name']

def battle_simulation(monster_a, monster_b):
    scores = {monster_a['name']: 0, monster_b['name']: 0}
    total_rolls = 601
    
    strength_modifier = calculate_strength_modifier(monster_a, monster_b)
    speed_modifier = calculate_speed_modifier(monster_a, monster_b)
    cunning_modifier = calculate_cunning_modifier(monster_a, monster_b)
    tenacity_modifier = calculate_tenacity_modifier(monster_a, monster_b)
    
    strengthMonster = monster_a if monster_a['Strength'] > monster_b['Strength'] else monster_b
    print(f"Strength Monster: {strengthMonster['name']}")
    
    strength_rerolls = []
    cunning_steals = []
    first_half_wins = 0
    tenacity_rolls = []
    
    for i in range(total_rolls):
        roll = random.randint(1, 6)
        winner = process_roll(roll, monster_a, monster_b, speed_modifier, cunning_modifier, strength_rerolls, cunning_steals)
        
        if i < total_rolls / 2 and winner == monster_a['name']:
            first_half_wins += 1
        
        if strength_modifier > 0:
            if roll == 3 and monster_b == strengthMonster:
                newRoll = random.randint(1, 6)
                strength_rerolls.append(1)
                if newRoll > 3:
                    scores[monster_b['name']] += 1
                else:
                    scores[monster_a['name']] += 1
                strength_modifier -= 1
            elif roll == 4 and monster_a == strengthMonster:
                newRoll = random.randint(1, 6)
                strength_rerolls.append(1)
                if newRoll < 4:
                    scores[monster_a['name']] += 1
                else:
                    scores[monster_b['name']] += 1
                strength_modifier -= 1
            else:
                scores[winner] += 1
        else:
            scores[winner] += 1
    
    for _ in range(tenacity_modifier):
        roll = random.randint(1, 6)
        winner = monster_a['name'] if roll <= 3 else monster_b['name']
        scores[winner] += 1
        tenacity_rolls.append((roll, winner))
    
    print("\nBattle Summary:")
    print(f"Total Strength Rerolls: {len(strength_rerolls)}")
    print(f"Total Cunning Steals: {len(cunning_steals)}")
    print(f"Total First Half Wins by {monster_a['name']}: {first_half_wins}")
    print(f"Tenacity Roll Outcomes: {tenacity_rolls}")
    
    return scores

monster_a = {"name": "StormBeast", "Speed": 9, "Strength": 5, "Cunning": 3, "Tenacity": 7}
monster_b = {"name": "IronDrake", "Speed": 5, "Strength": 9, "Cunning": 7, "Tenacity": 4}

print("Updated Monster Stat Blocks:")
print(monster_a)
print(monster_b)

battle_result = battle_simulation(monster_a, monster_b)

print("\nFinal Scores:")
print(battle_result)
```
